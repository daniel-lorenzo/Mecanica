__Se pide:__

a) Velocidad lineal crítica Vc-

b) Si se trabaja sin amortiguamiento (n=0) y al 70% de la velocidad crítica Vc calculada en el punto anterior, obtener la amplitud máxima.

c) Se decide acoplar un amortiguador para reducir la amplitud en un 41%. Determinar la constante de amortiguamiento viscoso C del amortiguador.

d) Identificar en el gráfico de transmisibilidad, de manera aproximada, los puntos de trabajo descritos en los puntos b) y c). (ver imágen adjunta "graf_trans")

## Solución

```python
# Datos:
M = 500    # [kg]
y_0 = 3e-2 # [m]
δ  = 10e-2 # [m]
d = 2      # [m]
g = 9.81   # [m/s^2]
```

```python
import numpy as np
```

### a)
Para calcular la frecuencia natural del sistema debemos, primero, calcular $k$.

$$\delta_{estatico} = \frac{g M_P}{k}$$

$$k = \frac{g M_P}{\delta_{estatico}}$$

```python
k = g*M/δ
```

```python
#@title k
print("k = %.1f [N/m]"%k)
```

Calculamos la frecuencia natural del sistema

$$p = \sqrt{ \frac{k}{M} }$$

```python
p = np.sqrt(k/M)
```

```python
#@title p
print("p = %.2f [1/s]"%p)
```

Para calcular la velocidad crítica planteamos $t = T_C$ (el pseudo periodo); $T_C = 2 \pi \omega$, pero en este caso puntual, podemos plantear:

$$T_C = \frac{2 \pi}{p}$$

```python
T_C = 2*np.pi/p
```

```python
#@title Tc
print("T_C = %.3f [s]"%T_C)
```

Calculando la velocidad crítica, usando el pseudo periodo:

$$v_c = \frac{d}{T_C}$$

```python
v_c = d/T_C
```

```python
#@title v_c
print("v_c = %.2f [m/s]"%v_c)
```

---
### b)
Para las siguientes condiciones:

```python
n = 0
v_b = 0.7*v_c
```

```python
#@title v_b
print("v_b = %.3f [m/s]"%v_b)
```

Para calcular la amplitud máxima, vamos a necesitar, primero, calcular $F_0$ y $\omega$

$$F_0 = k \cdot y_0$$

```python
F_0 = k*y_0
```

```python
#@title Fo
print("F_0 = %.1f [N]"%F_0)
```

Para calcular $\omega$, calculamos, primero, el nuevo pseudo periodo.

$$T_B = \frac{d}{v_b}$$

```python
T_B = d/v_b
```

```python
#@title T_B
print("T_B = %.2f [s]"%T_B)
```

$$\omega = \frac{2 \pi}{T_B}$$

```python
ω = 2*np.pi/T_B
```

```python
#@title ω
print("ω = %.2f [1/s]"%ω)
```

Al se un sistema amortiguado $n=0$ por lo tanto la $A_{max}$ se calcula como:

$$A_{max} = \frac{F_0/k}{\displaystyle \sqrt{\left( 1- \frac{\omega^2}{p^2} \right)^2}}$$

```python
A_max = (F_0/k)/np.sqrt(1 - ω**2/p**2)**2
```

```python
#@title A_max
print("A_max = %.4f [m]"%A_max)
```

---
### c) Como la amplitud se reduce un 41%, planteamos:

$$A = 0.59*$$

```python
A = 0.59*A_max
```

```python
#@title A
print("A = %.4f [m]"%A)
```

Conocida la nueva amplitud, y reemplazando en:

$$A = \frac{F_0}{k} \cdot \frac{1}{\displaystyle \sqrt{ \left( 1 - \frac{\omega^2}{p^2} \right)^2 + \left( 2 \frac{n}{n_c} \frac{\omega}{p} \right)^2 }}$$

Obtenenos:
$$\frac{n}{n_c} = 0,503$$

```python
n_nc = 0.503
n = 0.503*p
n_c = p
```

```python
#@title n
print("n = %.2f [1/s]"%n)
print("n_c = %.2f [1/s]"%n_c)
```

Calculamos, entonces, la constante de amortiguamiento:

```python
c = 2*n*M
```

```python
#@title C
print("c = %.1f [kg/s]"%c)
```
